### Ensemble Learning.

- Ensemble learning is supervised machine learning algorithm it is used to get the better accuracy by minimizing the variance during the period of training the data so that we will rid from overfitting.
- Enemble learning train the same data on the number of algorithm to the prediction and combine the all that result and on the basis of ,maxining votting for classification and avrage vote for the regression type we will decide final prediction. 
- The suggesting the ensemble that means it is collecting the all possible the result which comes from the diffrent diffrent algorithm and select the that result on the majority or avarage basis.This is known as ensemble learning.
- **In the ensemble learning we having the many more techiniques that are :-**

 - 1.Sequential Ensemble Learning i.e.Boosting techinique e.g.Adaboost classification.
 - 2.Parellel Ensemble learning techinique i.e. Bagging techinique e.g.RandomForestClassification and Regression.
 - 3.Staking ensemble learning  technique.i.e.max Voting techinique.

- Ensemble learning is usefull to maximum accuacy without getting overtting of our model wiith low training time.
- Because it uses the more than one algorithm to predict the better outcome or prediction.Every data is not always good for the evey algorithm to get good accuarcy to do that we need to use some bunch of algorithm to work on it so that we will highest accuarcy.Every model havinhg own advantage and disadvantage to overcome any weakness of the algorithm we use more than one together we will get the better accuarcy.

## Bagging.

- Bagging is one of the machine learning techique which is classification or regression.It having pattern of parellel training with the number of decison trees or we can it use the number of decision tresses to the get the possible outcome by combining the all result.
- After combining the all result it will use the aggregation on in such way so that we will get the result as max vote for classification and avrage vote for the regression type of model.
- It can be use for classifiction to the categorical clases or It can use for the regression for the continoue data.
- Bagging is helps to minimize the variance so that buden of the data is seperated to each model seperated in random fashion with replacement.

## Boosting.

- Boosting is nothing the Sequential type of ensemble learning technique in that it improving the performance of the classifiction step by step so we will get better accuarcy.
- In that ensemble we will always get the accuarcy better than previous model.
- In the boosting ensemble learning it will first train the data on the initial weight the predict the classification accuarcy if we get the low accuarcy then we update the weight of lower performing attribute i.e. Increses and descres the weight of higher performing attributes. and again train the model on the basis of updated attribute by doing so we we will make weak learning attribue to strong attribute.
- Finally if we chack it the accuarcy stuck somewhere that we can it is an optimal accuarcy that is found by doing the booosting ensemble learning technique.
